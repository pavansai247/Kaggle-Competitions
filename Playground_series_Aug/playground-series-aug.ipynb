```python
train_df = df.drop(labels = ["product_code","id","failure"],axis = 1)
test_dt = dt.drop(labels = ["product_code","id"],axis = 1)
y = df["failure"]
```

```python
y.head()
```

**Feature Engg**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x_df = train_df.apply(le.fit_transform)
x_dt = test_dt.apply(le.fit_transform)
x_df.head()
```

```python
from sklearn.impute import SimpleImputer
missing = SimpleImputer(missing_values=np.nan,strategy="median")
missing_df = missing.fit_transform(x_df)
missing_dt = missing.fit_transform(x_dt)
missing_df
```

```python
df_missing = pd.DataFrame(missing_df,index = train_df.index,columns = train_df.columns)
dt_missing = pd.DataFrame(missing_dt,index = test_dt.index,columns = test_dt.columns)
dt_missing.head()
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(df_missing)
ss.fit(dt_missing)
scaler_df = ss.transform(df_missing)
scaler_dt = ss.transform(dt_missing)
scaler_df
```

```python
df_scaler_df = pd.DataFrame(scaler_df,index=train_df.index,columns = train_df.columns)
dt_scaler_dt = pd.DataFrame(scaler_dt,index = test_dt.index,columns = test_dt.columns)
df_scaler_df
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca.fit(df_scaler_df)
pca.fit(dt_scaler_dt)
pca_df = pca.transform(df_scaler_df)
pca_dt = pca.transform(dt_scaler_dt)
pca_df
```

```python
x = pca_df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 20,test_size = 0.3)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train,y_train)
pred = lr.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train,y_train)
final = dtr.predict(x_test)
final
```

```python
accuracy_score(y_test,final)
```

```python
accuracy_score(y_train,dtr.predict(x_train))
```

```python
accuracy_score(y_train,lr.predict(x_train))
```

```python
pred0 = dtr.predict(pca_dt)
pred0
```

```python
pred1 = lr.predict(pca_dt)
pred1
```

```python
submission = pd.DataFrame({'ID':dt.id.values,'failure':pred1})
submission1 = pd.DataFrame({'ID':dt.id.values,'failure':pred0})
```
